### Processor Generation

```python
import re

def infer_gen_from_ghz(ghz):
    ghz = float(ghz)
    if ghz <= 2.0:
        return "5 Gen"
    elif 2.0 < ghz <= 2.5:
        return "6 Gen"
    elif 2.5 < ghz <= 2.8:
        return "8 Gen"
    elif 2.8 < ghz <= 3.1:
        return "9 Gen"
    elif 3.1 < ghz <= 3.3:
        return "11 Gen"
    elif ghz > 3.3:
        return "13 Gen"
    else:
        return "Unknown"

def extract_gen_or_estimate(cpu):
    if not isinstance(cpu, str):
        return "Unknown"

    # Try extracting gen from model number
    match = re.search(r'[iI][3579]\s?-?\s?(\d{4,5})', cpu)
    if match:
        number = match.group(1)
        if len(number) == 4:
            return f"{int(number[0])} Gen"  # e.g., 7200 -> 7 Gen
        elif len(number) == 5:
            return f"{int(number[:2])} Gen"  # e.g., 1135G7 -> 11 Gen

    # Fallback: Estimate from GHz
    freq_match = re.search(r'(\d\.\d+)\s?GHz', cpu)
    if freq_match:
        return infer_gen_from_ghz(freq_match.group(1))

    return "Unknown"

# Apply to DataFrame
laptop_df['CPU_Generation'] = laptop_df['Cpu'].astype(str).apply(extract_gen_or_estimate)

# Sample output
print(laptop_df[['Cpu', 'CPU_Generation']].head())
```

### Memory features

```python
# Flag SSD presence
laptop_df['Has_SSD'] = laptop_df['Memory'].astype(str).apply(lambda x: 1 if 'SSD' in x else 0)

# Total memory in GB
def extract_total_memory(memory):
    nums = re.findall(r'(\d+)', str(memory))
    return sum([int(n) for n in nums]) if nums else None

laptop_df['Total_Memory_GB'] = laptop_df['Memory'].apply(extract_total_memory)
laptop_df['Total_Memory_GB']
```

### Operating System Types

```python
def categorize_os(os):
    os = str(os).lower()
    if 'windows' in os:
        return 'Windows'
    elif 'mac' in os:
        return 'Mac'
    elif 'linux' in os:
        return 'Linux'
    else:
        return 'Other'

laptop_df['OS_Category'] = laptop_df['OpSys'].apply(categorize_os)
laptop_df['OS_Category']
```

# Hyphothesis Testing

## 1.Does the RAM size impacts the laptop Price ?

```python
# T - Test

# 1. RAM: High (>=8) vs Low (<8)
low_ram = laptop_df[laptop_df['Ram in GB'] < 8]['Price'].dropna()
high_ram = laptop_df[laptop_df['Ram in GB'] >= 8]['Price'].dropna()
tstat_ram, pval_ram = ttest_ind(high_ram, low_ram, equal_var=False)
pval_ram
```

# Inference

### Definitely the RAM impacts the  laptop price

## 2.Does the Operating System impacts the Price of the laptop ?

```python
from scipy.stats import f_oneway

# One Way ANOVA
# Group price data by OpSys
groups = laptop_df.groupby('OpSys')['Price'].apply(list)

# Perform one-way ANOVA
f_stat, p_value = f_oneway(*groups)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")
```

# Bivariate Analysis

```python
# Function to remove outliers based on IQR for each group (RAM)
def remove_outliers_iqr(df, col, group_col):
    def iqr_filter(group):
        Q1 = group[col].quantile(0.25)
        Q3 = group[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        return group[(group[col] >= lower_bound) & (group[col] <= upper_bound)]
    
    return df.groupby(group_col, group_keys=False).apply(iqr_filter)

# Apply to your dataset
laptop_df_cleaned = remove_outliers_iqr(laptop_df, 'Price', 'Ram in GB')
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Ram in GB', y='Price', data=laptop_df_cleaned)
plt.title('Price vs RAM (Outliers Removed)')
plt.show()
```
